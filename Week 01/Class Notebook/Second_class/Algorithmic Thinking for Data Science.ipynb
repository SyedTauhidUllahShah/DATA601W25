## <a name="complexity"></a>Complexity  

- The same problem can usually be solved via different algorithms that differ in efficiency. The term _computational complexity_ refers to the degree of difficulty of an algorithm.
  - _Time complexity_ characterizes the runtime.
  - _Space complexity_ characterizes the memory requirements.
  

- When analyzing algorithms, we are interested in counting the number of _primitive operations_ that the algorithm performs. This number is expressed as a function $f(n)$ where $n$ is the input size. 
  
  
- _Asymptotic complexity_ characterizes the growth rate of an algorithm for large input sizes $n$ (i.e. $n \rightarrow \infty$).
  - Big-O notation is commonly used to express the growth rate of algorithms.


### Example

- Suppose the time complexity of an algorithm is given by: $$f(n) = n^2 + 100n + 1000$$
  

- Here, as $n \rightarrow \infty$, the $n^2$ term dominates.
> We say that $f(n)$ is *$O(n^2)$*.
  


```python
def iterative_sum(n):
    total = 0
    for i in range(1, n + 1):  # Loop runs n times
        total += i
    return total

# Example usage
n = 10  # Large input size
result = iterative_sum(n)
print(f"Sum of numbers from 1 to {n} is: {result}")
```

## How much time does it take to execute?


```python
import time

def iterative_sum_1(n):
    start_time = time.time()  # Start timing
    total = 0
    for i in range(1, n + 1):  # Loop runs n times
        total += i
    end_time = time.time()  # End timing
    elapsed_time = end_time - start_time
    return total, elapsed_time

def iterative_sum_2(n):
    start_time = time.time()  # Start timing
    total = 0
    s = 0
    for i in range(1, n + 1): 
        s = i
        total += s
    end_time = time.time()  # End timing
    elapsed_time = end_time - start_time
    return total, elapsed_time

# Example usage
n = 1000000  # Large input size

result_1, time_1 = iterative_sum_1(n)
print(f"Sum of numbers from 1 to {n} using iterative_sum_1 is: {result_1} (calculated in {time_1:.6f} seconds)")

result_2, time_2 = iterative_sum_2(n)
print(f"Sum of numbers from 1 to {n} using iterative_sum_2 is: {result_2} (calculated in {time_2:.6f} seconds)")
```

### Linear Search Algorithm

**Linear search** is one of the simplest search algorithms. It involves traversing each element in a list sequentially until the desired element (target) is found, or the end of the list is reached.

---

### How It Works:
1. Start at the first element of the list.
2. Compare the current element with the target value:
   - If they match, return the index of the current element.
   - If they do not match, move to the next element.
3. Repeat until the target is found or the list is exhausted.
4. If the target is not found, return `-1`.

---

### Example:
Given the list `[3, 5, 2, 9, 7, 10]` and the target value `9`:
- Start comparing each element from the beginning:
  - Compare `3` with `9` → Not a match.
  - Compare `5` with `9` → Not a match.
  - Compare `2` with `9` → Not a match.
  - Compare `9` with `9` → Match found!
- Return the index `3`.

---

### Time Complexity:
- **Best Case**: \( O(1) \) — The target is found at the first element.
- **Worst Case**: \( O(n) \) — The target is not found, or it is the last element in the list.

### Space Complexity:
- \( O(1) \): Linear search requires no additional memory apart from a few variables.



```python
def linear_search(arr, target):
    """
    Searches for the target value in the array using linear search.
    
    Args:
        arr (list): The list to search in.
        target (any): The value to search for.
        
    Returns:
        int: The index of the target if found, else -1.
    """
    for index in range(len(arr)):
        if arr[index] == target:
            return index  # Return the index if the target is found
    return -1  # Return -1 if the target is not found

# Example usage
numbers = [3, 5, 2, 9, 7, 10]
target = 9

result = linear_search(numbers, target)
if result != -1:
    print(f"Element {target} found at index {result}.")
else:
    print(f"Element {target} not found.")
```

### Big-O Observations

Here are some useful properties that you should know:

- Constants get _absorbed_ by the notation.

  E.g. $f(n) = 2n^2 + 3n + 1 = O(2n^2) = O(n^2)$. 
  We usually characterize growth rates without leading constants.
  
  
- We can use big-O to characterize the growth rate of lower order terms. 

  E.g. $f(n) = n^2 + 100n + 1000 = n^2 + O(n)$. 
  
  
- Big-O is imprecise about the tightness of the upper bound.

  E.g. $f(n) = 2n^2$ is $O(n^2)$ but also $O(n^3)$, $O(n^4)$ etc.
  We usually want the tightest upper bound.

### Common Classes of Algorithms

In order of *increasing* complexity:

    
| **Class** | constant | logarithmic | linear |  &nbsp;&nbsp; n log n &nbsp; &nbsp; | quadratic | &nbsp; cubic &nbsp; | exponential | factorial |
| --- | --- | --- | --- | --- | --- | --- | --- | --- |
| **Complexity** | $O(1)$ | $O(\lg n)$ | $O(n)$ | $O(n \lg n)$ | $O(n^2)$ | $O(n^3)$ | $O(2^n)$ | $O(n!)$ |



### Bubble Sort Algorithm

**Bubble Sort** is a simple sorting algorithm that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. This process is repeated until the list is sorted.

Input = [64, 34, 25, 12, 22, 11, 90]


Output = Sorted list: [11, 12, 22, 25, 34, 64, 90]




 (https://favtutor.com/blogs/bubble-sort-python)

```python
def bubble_sort(arr):
    """
    Sorts an array using the bubble sort algorithm.
    
    Args:
        arr (list): The list of elements to sort.
        
    Returns:
        list: The sorted list.
    """
    n = len(arr)
    for i in range(n):
        # Flag to track if any swaps were made during this pass
        swapped = False
        for j in range(0, n - i - 1):  # Reduce the range as the largest elements bubble to the end
            if arr[j] > arr[j + 1]:
                # Swap if the current element is greater than the next
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # If no swaps were made, the array is already sorted
        if not swapped:
            break
    return arr

# Example usage
numbers = [64, 34, 25, 12, 22, 11, 90]
sorted_numbers = bubble_sort(numbers)
print(f"Sorted list: {sorted_numbers}")
```

### Bubble Sort Time Complexity:

| **Case**        | **Time Complexity** | **Explanation**                                                         |
|------------------|---------------------|-------------------------------------------------------------------------|
| **Best Case**    | \( O(n) \)          | Only one pass needed, no swaps required (array is already sorted).      |
| **Worst Case**   | \( $O(n^2)$ \)        | Requires \( $n \times (n-1)/2$ \) comparisons (array is sorted in reverse).|
| **Average Case** | \( $O(n^2)$ \)        | Most inputs require significant comparisons and swaps.                  |

### What is Binary Search?

**Binary search** is an efficient algorithm for finding the position of a target value within a sorted array. It works by repeatedly dividing the search interval in half, reducing the search space with each step.

---

### How Binary Search Works

#### 1. **Initial Setup**:
- Define two pointers:
  - `left` (start of the array).
  - `right` (end of the array).
- Calculate the middle index:
  \[
  \text{mid} = \frac{\text{left} + \text{right}}{2} \, \text{(integer division)}.
  \]

#### 2. **Comparison**:
- **If** the middle value (`arr[mid]`) equals the target:
  - The search is successful, and the target's index is returned.
- **If** the target is smaller than the middle value:
  - Narrow the search to the left half by updating `right = mid - 1`.
- **If** the target is larger than the middle value:
  - Narrow the search to the right half by updating `left = mid + 1`.

#### 3. **Repeat**:
- Continue dividing the array and comparing until:
  - The target is found.
  - Or the search interval becomes empty (`left > right`).

#### 4. **Output**:
- **If the target is found**:
  - Return its index.
- **If the target is not in the array**:
  - Return `-1`.




```python
def binary_search(arr, target):
    """
    Perform binary search on a sorted array to find the target value.
    Returns the index of the target if found, otherwise -1.
    """
    left, right = 0, len(arr) - 1  # Define the search boundaries

    while left <= right:
        mid = (left + right) // 2  # Calculate the middle index
        mid_value = arr[mid]       # Value at the middle index

        if mid_value == target:    # Target found
            return mid
        elif mid_value < target:   # Target is in the right half
            left = mid + 1
        else:                      # Target is in the left half
            right = mid - 1

    return -1  # Target not found
```

### Input Size and Halving

Binary Search starts with the entire array of size $n$. At each step, the algorithm halves the search range:

1. **Initially**:
   $ \text{Size of range} = n $
2. **After 1 step**:
   $ \text{Size of range} = \frac{n}{2} $
3. **After 2 steps**:
   $ \text{Size of range} = \frac{n}{4} $
4. **After $k$ steps**:
   $ \text{Size of range} = \frac{n}{2^k} $

---

### Termination Condition

Binary Search stops when the range is reduced to **1 element**:

$ \frac{n}{2^k} = 1 $

---

### Solving for $k$

1. Rearrange the equation:
   $ 2^k = n $
2. Take the logarithm (base 2) of both sides:
   $ k = \log_2(n) $

This means the algorithm performs $ \log_2(n) $ steps in the **worst case**.

---

### Time Complexity

#### 1. **Work Per Step**:
- In each step, Binary Search performs:
  - One comparison to check if the middle element matches the target.
  - One decision to move left or right.
- These operations take **constant time**:
  $ O(1) $

#### 2. **Total Steps**:
- The total number of steps is:
  $ k = \log_2(n) $

#### 3. **Overall Complexity**:
- Multiply the work per step ($ O(1) $) by the number of steps ($ \log_2(n) $):
  $ O(1) \cdot \log_2(n) = O(\log n) $

Thus, the time complexity of Binary Search is:

$ O(\log n) $

### What is Merge Sort?

**Merge Sort** is a divide-and-conquer sorting algorithm. It works by recursively dividing the input array into two halves, sorting ### Problem:
Sort the list  
`[8, 3, 1, 6, 7, 2, 5, 4]` using merge sort.

---

### Step-by-Step Execution:

#### **Divide Phase:**

Split the list into halves repeatedly until each subarray contains only 1 element.  
Example splits:  
- `[8, 3, 1, 6, 7, 2, 5, 4] → [8, 3, 1, 6]` and `[7, 2, 5, 4]`
- `[8, 3, 1, 6] → [8, 3]` and `[1, 6]`
- `[8, 3] → [8]` and `[3]`  
Similarly, for the right half:  
- `[7, 2, 5, 4] → [7, 2]` and `[5, 4]`
- `[7, 2] → [7]` and `[2]`  
- `[5, 4] → [5]` and `[4]`

---

#### **Conquer (Merge Phase):**

Merge the subarrays back together in sorted order:
1. Merge `[8]` and `[3]` to get `[3, 8]`.
2. Merge `[1]` and `[6]` to get `[1, 6]`.
3. Merge `[3, 8]` and `[1, 6]` to get `[1, 3, 6, 8]`.

For the right half:
1. Merge `[7]` and `[2]` to get `[2, 7]`.
2. Merge `[5]` and `[4]` to get `[4, 5]`.
3. Merge `[2, 7]` and `[4, 5]` to get `[2, 4, 5, 7]`.

Finally:
1. Merge `[1, 3, 6, 8]` and `[2, 4, 5, 7]` to get `[1, 2, 3, 4, 5, 6, 7, 8]`.

---

### **Final Sorted List:**

`[1, 2, 3, 4, 5, 6, 7, 8]`

---

### **Why Is This $O(n \log n)$?**

#### **Number of Levels (Divide Phase):**
The array is repeatedly split in half until we have subarrays of size 1.  
The number of splits is $\log_2(n)$, where $n$ is the size of the array.

#### **Work per Level (Merge Phase):**
At each level, we merge all elements in the array.  
Merging takes $$O(n)$$ time because each element is compared and placed in the correct order.

#### **Total Work:**
Multiply the work per level ($O(n)$) by the number of levels ($O(\log n)$):  
$O(n) \cdot O(\log n) = O(n \log n)$

### What is Recursion?

**Recursion** is a programming technique where a function calls itself to solve smaller instances of the same problem. It is commonly used to solve problems that can be broken down into similar subproblems.

---

### Key Components of Recursion

1. **Base Case**:
   - A condition that stops the recursion. Without a base case, recursion will continue indefinitely, leading to a stack overflow.
   - Example: In a factorial function, the base case is when $n = 0$ or $n = 1$.

2. **Recursive Case**:
   - The part of the function where it calls itself with a smaller input, progressively working toward the base case.



```python
# Example: Factorial 
# Recall the recursive definition of the factorial: n! = n * (n-1)!
#                                                   0! = 1
# We can code this up concisely as follows:

def fact(n):
    return 1 if n == 0 else n * fact(n-1)

# Let's also plot it to see how it grows in comparison to linear and 
# exponential growth

import matplotlib.pyplot as plt
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = range(1,21)
plt.semilogy( x, [fact(n) for n in x] ,'k.-', label='$n!$')
plt.semilogy( x, [n for n in x] ,'r.-', label='$n$')
plt.semilogy( x, [2**n for n in x] ,'b.-', label='$2^n$')
plt.legend();
#fig.set_size_inches(8,6)
plt.show()
```

## <a name="search_and_sort"></a>Searching and Sorting Algorithms

- Sorting and searching are fundamental operations and most libraries provide general purpose implementations of efficient sorting and searching algorithms.


- It is good to know what's going on under the hood so that you can use the algorithm that's best suited for the job.


- For _ordered_ data, _binary search_ is more efficient compared to _sequential_ search.


- For comparison based sorting, avoid naïve quadratic algorithms. 


- Check the complexities of various operations in Python at https://wiki.python.org/moin/TimeComplexity.  

### Dictionaries

- Dictionaries are specially designed to perform searches efficiently. Using hash tables, Python `dict` and `set` data structures have an average case complexity of $O(1)$ (constant time).


- Dictionaries sacrifice data ordering, there is no guarantee that data will be in any particular order. 

- You can sort keys and values in a `dict` separately as lists. 

### Sorting

- *Insertion sort*, *selection sort* and *bubble sort* have an expected time complexity of $O(n^2)$.


- More sophisticated sorting algorithms such as *merge sort* and *quick sort* are recursive and reduce the expected case complexity to $O(n \lg n)$.

- Python's list sorting functions (`list.sort()` and `sorted()`) use a hybrid sorting algorithm that performs quite well and has an expected time complexity of $O(n \lg n)$.


- NumPy provides more control by allowing the programmer to specify the sorting algorithm.
