```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Task 1: Reading Data with NumPy
file_path = "movies.txt"
data = np.genfromtxt(file_path, delimiter="\n", dtype=str, encoding="utf-8")

# Task 2: Data Preprocessing
# Strip whitespace and convert to lowercase
movies = np.array([movie.strip().lower() for movie in data if movie.strip()])

# Task 3: Analyzing the Data
# Total number of movie entries
total_movies = len(movies)
print(f"Total number of movies: {total_movies}")

# Number of unique movie titles
unique_movies = np.unique(movies)
num_unique_movies = len(unique_movies)
print(f"Number of unique movies: {num_unique_movies}")

# Duplicate entries and their frequencies
movie_counts = Counter(movies)
duplicates = {movie: count for movie, count in movie_counts.items() if count > 1}
print("Duplicate movies:")
for movie, count in duplicates.items():
    print(f"{movie}: {count}")

# Frequency distribution
frequency_distribution = Counter(movie_counts.values())

# Task 4: Data Visualization with Matplotlib
# 4.1 Bar chart for the top 5 most frequent movies
top_5 = movie_counts.most_common(5)
movies_top_5, counts_top_5 = zip(*top_5)

plt.figure(figsize=(8, 6))
plt.bar(movies_top_5, counts_top_5, color="grey")
plt.title("Top 5 Most Frequent Movies")
plt.xlabel("Movie Titles")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 4.2 Pie chart for unique vs repeated movies
num_repeated = total_movies - num_unique_movies
unique_vs_repeated = [num_unique_movies, num_repeated]
labels = ["Unique", "Repeated"]

plt.figure(figsize=(6, 6))
plt.pie(unique_vs_repeated, labels=labels, autopct="%1.1f%%", colors=["grey", "skyblue"], startangle=140)
plt.title("Proportion of Unique vs Repeated Movies")
plt.axis("equal")  # Ensure the pie chart is a circle
plt.show()

# Task 5: Advanced Visualization
# Histogram for distribution of frequencies
frequencies, counts = zip(*frequency_distribution.items())

plt.figure(figsize=(8, 6))
plt.bar(frequencies, counts, width=0.8, color="blue", alpha=0.7)
plt.title("Distribution of Movie Frequencies")
plt.xlabel("Frequency of Appearance")
plt.ylabel("Number of Movies")
plt.xticks(frequencies)
plt.tight_layout()
plt.show()
```
