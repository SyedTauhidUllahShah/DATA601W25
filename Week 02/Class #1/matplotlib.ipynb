## 3. Basic Plotting <a name="basic-plotting"></a>

Matplotlib's pyplot module provides a simple interface for creating plots.

### Purpose of Line Plots
Line plots are used to display trends over a continuous range, such as time series data or functions. They are useful for visualizing how a variable changes over time or in response to another variable.


````python
import numpy as np


# Generate some data using NumPy
x = np.linspace(0, 10, 100)       # 100 points from 0 to 10
y = np.sin(x)                     # sine of x

# Create a line plot
plt.figure(figsize=(8, 4))        # optional: set figure size
plt.plot(x, y)                    # plot y vs. x
plt.title("Simple Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)                    # add grid lines
plt.show()
````

## 4. Plot Customization <a name="customization"></a>

Customize plots by adjusting colors, styles, markers, and more.

````python
# Using different line styles and markers
plt.figure(figsize=(8, 4))
plt.plot(x, y, color='red', linestyle='--', marker='o', label='sin(x)')
plt.title("Customized Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()      # display legend
plt.show()
````

## 5. Multiple Plots and Subplots <a name="subplots"></a>

Subplots allow you to display multiple related plots in one figure for comparison. They are useful when visualizing different aspects of data side by side.

````python
# Create two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) #axes is a NumPy array-like object where each element corresponds to an individual subplot (e.g., axes[0] and axes[1]).
#fig: The figure object that contains all subplots.

axes[0].plot(x, np.sin(x), 'b-', label='sin(x)')
axes[0].set_title("Sine Wave")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")
axes[0].legend()

axes[1].plot(x, np.cos(x), 'r-', label='cos(x)')
axes[1].set_title("Cosine Wave")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")
axes[1].legend()

plt.tight_layout()  # Adjusts subplot params for a neat layout
plt.show()
````

## 6. Plotting Different Chart Types <a name="chart-types"></a>

Matplotlib can create various chart types. Each serves a specific purpose:
- **Scatter Plots:** Show relationships between two variables. Ideal for correlation and distribution studies.
- **Density Plots:** Density charts (also called density plots or 2D histograms) visualize the distribution of data across one or two dimensions.
- **Bar Charts:** Compare quantities across categories. Common in categorical data analysis.
- **Histograms:** Display distribution of numerical data. Useful for understanding data distribution and outliers.
- **Pie Charts:** Show proportions of a whole. Best for categorical data with few categories.

Let's explore each with examples.

### Scatter Plot

````python
# Scatter plot: random points
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(6, 6))
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.8)
plt.title("Random Scatter Plot")
plt.colorbar()  # show color scale
plt.show()
````

````python
# Generate random data
x = np.random.randn(10000)
y = 2 * x + np.random.randn(10000)

# Create a figure and axes
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Scatter plot
axes[0].scatter(x, y, alpha=0.5)
axes[0].set_title("scatter plot")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")

# Density plot
hist = axes[1].hist2d(x, y, bins=100, cmap='viridis')
axes[1].set_title("density plot")
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Y-axis")

# Add colorbar for density plot
fig.colorbar(hist[3], ax=axes[1], orientation='horizontal')

# Show the plots
plt.tight_layout()
plt.show()
````

### Bar Chart

**Purpose:** Bar charts compare discrete categories. They are used in situations where you want to compare quantities across different groups or categories.

````python
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=5)

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
````

### Histogram

**Purpose:** Histograms display the distribution of a dataset by grouping values into bins. They are used to see patterns such as central tendency, dispersion, skewness, and outliers.

````python
data_hist = np.random.randn(1000)  # normal distribution

plt.figure(figsize=(6, 4))
plt.hist(data_hist, bins=30, color='black', alpha=0.7) # a bin refers to the range of values into which data points are grouped. 
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

### Pie Chart

**Purpose:** Pie charts show the proportional composition of a whole for categorical data. They are best used when there are a limited number of categories.

````python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # Moves slices outward according to the explode values.

plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=140)
plt.title("Pie Chart")
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()
````

## 7. Working with NumPy Data <a name="numpy-integration"></a>

Matplotlib works seamlessly with NumPy arrays. Here are some examples:

````python
# Generate data using NumPy
x_num = np.linspace(0, 2*np.pi, 400)
y_num = np.sin(x_num**2)

# Plot using Matplotlib
plt.figure(figsize=(8, 4))
plt.plot(x_num, y_num)
plt.title("Plot of sin(x^2)")
plt.show()
````

## 8. Styling Plots <a name="styling"></a>

Matplotlib offers various ways to style plots to make them visually appealing.

````python
plt.style.use('ggplot')  # Use a pre-defined style

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x^2)')
plt.title("Styled Plot")
plt.xlabel("x")
plt.ylabel("sin(x^2)")
plt.legend()
plt.show()

# Reset style to default
plt.style.use('default')
````

## 10. 3D Plotting <a name="3d-plotting"></a>

Matplotlib supports 3D plotting via `mpl_toolkits.mplot3d`.

````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Create data for a 3D parametric curve
t = np.linspace(0, 10, 100)
x3d = np.sin(t)
y3d = np.cos(t)
z3d = t

ax.plot(x3d, y3d, z3d, label='3D curve')
ax.set_title("3D Line Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
ax.legend()

plt.show()
````

## 11. Saving Figures <a name="saving"></a>

Save plots to files for later use:

````python
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Saving This Plot")
plt.savefig("sine_wave.png")  # Save as PNG
plt.savefig("sine_wave.pdf")  # Save as PDF
plt.close()  # Close the figure to free memory
````
